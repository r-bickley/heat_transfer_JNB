```python
#Project 1, Part 1
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
#Define Prescribed Temperatures
T_t = 1000
T_b = 350
```

```python
#Define the fin dimensions
w = 0.1
l = 0.1
```

```python
#Define the number of x & y divisions
xdiv = 4
ydiv = 4
#Define linear spaces for x and y
x = np.linspace(0,l,xdiv)
y = np.linspace(0,w,ydiv)
```

```python
#Make the empty matrix
theta = np.zeros(shape = (xdiv,ydiv))
```

```python
#Populate the matrix iteratively
for i in range(ydiv):
    for j in range(xdiv):
        for n in range(1,100):
            theta[i,j] += (2/m.pi)*((((-1)**(n+1))+1)/n)*(np.sin(n*m.pi*x[j]/l))*((np.sinh(m.pi*n*y[i]/l))/(np.sinh(n*m.pi*w/l)))
```

```python
#Calculate the temperatures based on the theta matrix
T = T_b+(T_t-T_b)*theta
```

```python
#Make a 2D Contour Plot
plt.contour(x,y,T)
ax = plt.gca()
ax.set_aspect('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.colorbar()
plt.title('Analytical Solution')
```
